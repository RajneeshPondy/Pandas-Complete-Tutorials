### Bar Graph

```python
google = data.DataReader(name= "GOOG", data_source="yahoo", start= "2016-1-01", end="2016-12-31")
google.head(3)
```

```python
def rank_performanse(stock_price):
    if stock_price <= 700 :
        return "Poor"
    elif stock_price > 700 and stock_price <= 750:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google["Close"].apply(rank_performanse)
```

```python
google["Close"].apply(rank_performanse).value_counts()
```

```python
google["Close"].apply(rank_performanse).value_counts().plot()
```

```python
google["Close"].apply(rank_performanse).value_counts().plot(kind="line")
```

```python
google["Close"].apply(rank_performanse).value_counts().plot(kind = "bar")
```

```python
google["Close"].apply(rank_performanse).value_counts().plot(kind ="barh")
```

### Pie Charts

```python
apple = data.DataReader(name= "AAPL", data_source="yahoo", start= "2012-1-01", end="2016-12-31")
apple.head(3)
```

```python
apple["Close"].mean()
```

```python
def rank_performanse(stock_price):
    if stock_price >= 93.3617192254726:
        return "Above Average"
    else:
        return "Below Average"
    
```

```python
plt.style.use("ggplot")
apple["Close"].apply(rank_performanse).value_counts().plot(kind = "pie", legend = True)
```

### Histograms

```python
google = data.DataReader(name= "GOOG", data_source="yahoo", start= "2004-01-01", end="2016-12-31")
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
google["High"].apply(custom_round).value_counts().sort_index()
```

```python
google["High"].apply(custom_round).nunique()
```

```python
google["High"].apply(custom_round).plot(kind = "hist")
```

```python
google["High"].apply(custom_round).plot(kind = "hist", bins= 9)
```
