### Outer Joins

```python
week1 = pd.read_csv('data/Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('data/Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('data/Restaurant - Customers.csv')
foods = pd.read_csv('data/Restaurant - Foods.csv')
```

```python
week1.head(3)
```

```python
week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1"," - Week 2"])
```

```python
merged = week1.merge(week2, how='outer', on='Customer ID', suffixes=[" A"," B"],
            indicator=True)
```

```python
merged["_merge"].value_counts()
```

```python
mask = merged["_merge"].isin(["left_only","right_only"])
merged[mask]
```

```python
mask2 = merged['_merge'].isin(["both"])
merged[mask2]
```

### Merging by Indexes with the left_index and right_index Parameter

```python
week1 = pd.read_csv('data/Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('data/Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('data/Restaurant - Customers.csv', index_col="ID")
foods = pd.read_csv('data/Restaurant - Foods.csv', index_col="Food ID")
```

```python
customers.head(3)
```

```python
foods.head(3)
```

```python
sales = week1.merge(customers, how="left", left_on="Customer ID", right_index=True)
```

```python
sales.head(2)
```

```python
foods.head(3)
```

```python
sales.merge(foods, how="left", left_on="Food ID", right_index=True)
```
