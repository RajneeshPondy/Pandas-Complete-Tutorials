### The .transpose() Method on MultiIndex

```python
bigmac = pd.read_csv('data/bigmac.csv', parse_dates=["Date"], index_col=["Date","Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.transpose()
```

```python
bigmac = bigmac.transpose()
```

```python
bigmac.shape
```

```python
bigmac.head(1)
```

```python
bigmac.index
```

```python
bigmac.ix["Price in US Dollars"]
```

```python
bigmac.ix["Price in US Dollars",("2016-01-01","China")]
```
